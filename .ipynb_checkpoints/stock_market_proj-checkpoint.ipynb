### To find std (volatility) for a stock, first find % return for each year for each stock, then do std(% return for the 5 years)

```python
stock_df.head()
```

```python
stock_df['year'] = pd.DatetimeIndex(stock_df['date']).year
stock_df.tail()
```

```python
stock_date_range = stock_df.groupby(['Name','year'])['date'].agg([min, max]).reset_index()
stock_date_range.head()
```

```python
result = pd.merge(stock_date_range, stock_df[['date','close','Name']], how='inner', 
                  left_on=  ['Name', 'min'],
                    right_on= ['Name', 'date'])

result = result[['Name','year','min','max','close']]
result = result.rename(columns={'close':'min_close'})
result.head()
```

```python
result = pd.merge(result, stock_df[['date','close','Name']], how='inner', 
                  left_on=  ['Name', 'max'],
                    right_on= ['Name', 'date'])

result = result[['Name','year','min','max','min_close','close']]
result = result.rename(columns={'close':'max_close'})
result.head()
```

```python
result['percentreturn'] = (result['max_close'] - result['min_close'])/result['min_close'] * 100.0
result.head(n=20)
```

```python
result.groupby('Name')['percentreturn'].mean().head()
```

```python
std_result = result.groupby('Name')['percentreturn'].std().reset_index()
std_result.describe()
```

```python
std_result.sort_values('percentreturn').head()
```

```python
std_result.sort_values('percentreturn').tail(n=10)
```

```python
# check the 6 stocks with NaN, they are all stocks that do not have the full 5 years of data
result[result['Name'].isin(['APTV','BHF','BHGE','DWDP','DXC','HLT'])]
```

```python
# only consider stocks that have the full 5 years of data
std_result = std_result[std_result['percentreturn'].notna()]
print(std_result.sort_values('percentreturn').head())
print(std_result.sort_values('percentreturn').tail())
```

### goal: find stocks based on high vs low volatility, and look at their highest percentage gain differences
<ui><li>sort the stocks in both categories by volatility</li>
<li>graph the most volatile vs least volatile in each category (measure volatility by StDev) over 5 yrs time --> 2 graphs</li>
<li>for those under 25% percentile of volatility, graph the highest percentage gain over 5 yrs time --> 2 graphs</li>
<li>for those above 75% percentile of volatility, graph the highest percentage gain over 5 yrs time --> 2 graphs</li>
<li>calc diff between the highest percentage gains of the high vs low volatility stocks for each category --> 2 numbers</li></ui>

```python
cheap_std = std_result[std_result['Name'].isin(cheap_stock)].sort_values('percentreturn')
cheap_std
```

```python
# find stocks that's 'cheap' and has volatility (as defined by std) below 25 percentile
cheap_lowV_n = len(cheap_std) // 4
cheap_lowV = cheap_std[0:cheap_lowV_n]
cheap_lowV
```

```python
# find stocks that's 'cheap' and has volatility (as defined by std) above 75 percentile
cheap_hiV_n = len(cheap_std) // 4
cheap_hiV = cheap_std[-cheap_lowV_n:len(cheap_std)]
cheap_hiV
```

```python
# check for the earliest and latest stock price entry for the cheap stocks of interests
c_date_range = stock_df[(stock_df['Name'].isin(cheap_lowV.Name)) | (stock_df['Name'].isin(cheap_hiV.Name))].groupby('Name')['date'].agg(['min', 'max'])
c_date_range = c_date_range.reset_index()
c_date_range
```

```python
def percent_gain_dict(stk_names, date_range_df):

    pg_dict = {}
    for s in stk_names:   
        min_date = date_range_df[date_range_df['Name'] == s]['min']
        max_date = date_range_df[date_range_df['Name'] == s]['max']
        s_begin = float(stock_df[(stock_df['Name'] == s) & (stock_df['date'].isin(min_date))]['close'])
        s_end = float(stock_df[(stock_df['Name'] == s) & (stock_df['date'].isin(max_date))]['close'])
        pg_dict[s] = (s_end - s_begin) / s_begin

    return pg_dict
```

```python
# 'BSX' is the "cheap stock" with relatively low volatility and highest percentage gain over 5 yrs
cl_dict = percent_gain_dict(cheap_lowV.Name, c_date_range)
cl_dict
sorted_cl_dict = sorted(cl_dict.items(), key=lambda x: x[1], reverse=True)
sorted_cl_dict
```

```python
# 'MU' is the cheap stock with relatively high volatility and highest percentage gain over 5 yrs
ch_dict = percent_gain_dict(cheap_hiV.Name, c_date_range)

sorted_ch_dict = sorted(ch_dict.items(), key=lambda x: x[1], reverse=True)
sorted_ch_dict
```

```python
# diff of percentage gain over 5 yrs btwn highest performing high vs low volatility stocks for cheap stocks
cheap_percent_gain_diff = sorted_ch_dict[0][1] - sorted_cl_dict[0][1]
cheap_percent_gain_diff
```

```python
pricy_std = std_result[std_result['Name'].isin(pricy_stock)].sort_values('percentreturn')
pricy_std.head()
```

```python
# find stocks that's 'pricy' and has volatility (as defined by std) below 25 percentile
pricy_lowV_n = len(pricy_std) // 4
pricy_lowV = pricy_std[0:pricy_lowV_n]

# find stocks that's 'pricy' and has volatility (as defined by std) above 75 percentile
pricy_hiV_n = len(pricy_std) // 4
pricy_hiV = pricy_std[-pricy_lowV_n:len(pricy_std)]

# check for the earliest and latest stock price entry for the pricy stocks of interests
p_date_range = stock_df[(stock_df['Name'].isin(pricy_lowV.Name)) | (stock_df['Name'].isin(pricy_hiV.Name))].groupby('Name')['date'].agg(['min', 'max'])
p_date_range = p_date_range.reset_index()
p_date_range.head()
```

```python
len(pricy_stock)//4*2
```

```python
print(p_date_range['min'].unique())
print(p_date_range['max'].unique())
```

```python
# 'UNH' is the pricy stock with relatively low volatility and highest percentage gain over 5 yrs
pl_dict = percent_gain_dict(pricy_lowV.Name, p_date_range)

sorted_pl_dict = sorted(pl_dict.items(), key=lambda x: x[1], reverse=True)
sorted_pl_dict[0:5]
```

```python
# 'NVDA' is the pricy stock with relatively high volatility and highest percentage gain over 5 yrs
ph_dict = percent_gain_dict(pricy_hiV.Name, p_date_range)

sorted_ph_dict = sorted(ph_dict.items(), key=lambda x: x[1], reverse=True)
sorted_ph_dict[0:5]
```

```python
# diff of percentage gain over 5 yrs btwn highest performing high vs low volatility stocks for pricy stocks
pricy_percent_gain_diff = sorted_ph_dict[0][1] - sorted_pl_dict[0][1]
pricy_percent_gain_diff
```

```python
# diff of percentage gain over 5 yrs btwn highest performing pricy vs cheap stocks, for each volatility category
low_v_percent_gain_diff = sorted_pl_dict[0][1] - sorted_cl_dict[0][1]
high_v_percent_gain_diff = sorted_ph_dict[0][1] - sorted_ch_dict[0][1]
print(low_v_percent_gain_diff)
print(high_v_percent_gain_diff)
```
