```python
# to find std (volatility), need to find % return for each year for each stock, then do std(% return for the 5 years)
```

```python
stock_df.head()
```

```python
stock_df['year'] = pd.DatetimeIndex(stock_df['date']).year
stock_df.tail()
```

```python
stock_date_range = stock_df.groupby(['Name','year'])['date'].agg([min, max]).reset_index()
stock_date_range.head()
```

```python
result = pd.merge(stock_date_range, stock_df[['date','close','Name']], how='inner', 
                  left_on=  ['Name', 'min'],
                    right_on= ['Name', 'date'])

result = result[['Name','year','min','max','close']]
result = result.rename(columns={'close':'min_close'})
result.head()
```

```python
result = pd.merge(result, stock_df[['date','close','Name']], how='inner', 
                  left_on=  ['Name', 'max'],
                    right_on= ['Name', 'date'])

result = result[['Name','year','min','max','min_close','close']]
result = result.rename(columns={'close':'max_close'})
result.head()
```

```python
result['percentreturn'] = (result['max_close'] - result['min_close'])/result['min_close'] * 100.0
result.head(n=20)
```

```python
result.groupby('Name')['percentreturn'].mean().head()
```

```python
std_result = result.groupby('Name')['percentreturn'].std().reset_index()
std_result.describe()
```

```python
std_result.head()
```

```python
cheap_std = std_result[std_result['Name'].isin(cheap_stock)].sort_values('percentreturn')
cheap_std
```

```python
# find stocks that's 'cheap' and has volatility (as defined by std) below 25 percentile
cheap_lowV_n = len(cheap_std) // 4
cheap_lowV = cheap_std[0:cheap_lowV_n]
cheap_lowV
```

```python
# find stocks that's 'cheap' and has volatility (as defined by std) above 75 percentile
cheap_hiV_n = len(cheap_std) // 4
cheap_hiV = cheap_std[-cheap_lowV_n:len(cheap_std)]
cheap_hiV
```

```python
# check for the earliest and latest stock price entry for the cheap stocks of interests
c_date_range = stock_df[(stock_df['Name'].isin(cheap_lowV.Name)) | (stock_df['Name'].isin(cheap_hiV.Name))].groupby('Name')['date'].agg(['min', 'max'])
c_date_range
```

```python
# 'RF' is the 'cheap stock' with relatively low volatility and highest percentage gain
# Need to fix starting here#

cl_dict = {}
for i, cl in cheap_lowV.iterrows():   
    cl_begin = stock_df[(stock_df['Name'] == cl['Name']) & (stock_df['date'] == '2013-02-08')]['close']
    cl_end = stock_df[(stock_df['Name'] == cl['Name']) & (stock_df['date'] == '2018-02-07')]['close']
    cl_dict[cl['Name']] = (cl_end - cl_begin) / cl_begin

cl_dict
```

```python
# 'MU' is the 'cheap stock' with relatively high volatility and highest percentage gain

ch_dict = {}
for cl in cheap_hiV.Name:   
    cl_begin = float(stock_df[(stock_df['Name'] == cl) & (stock_df['date'] == '2013-02-08')]['close'])
    cl_end = float(stock_df[(stock_df['Name'] == cl) & (stock_df['date'] == '2018-02-07')]['close'])
    ch_dict[cl] = (cl_end - cl_begin) / cl_begin

ch_dict
```

```python
# diff of percentage gain over 5 yrs btwn highest performing high vs low volatility stocks
cheap_percent_gain_diff = ch_dict['MU'] - cl_dict['RF'] 
cheap_percent_gain_diff
```

```python

pricy_std = stock_df[stock_df['Name'].isin(pricy_stock)].groupby('Name')['close'].std().sort_values()

# find stocks that's 'pricy' and has volatility (as defined by std) below 25 percentile
pricy_lowV_n = len(pricy_std) // 4
pricy_lowV = pricy_std[0:pricy_lowV_n]

# find stocks that's 'pricy' and has volatility (as defined by std) above 75 percentile
pricy_hiV_n = len(pricy_std) // 4
pricy_hiV = pricy_std[-pricy_lowV_n:len(pricy_std)]

# check for the earliest and latest stock price entry for the pricy stocks of interests
pl_date_range = stock_df[stock_df['Name'].isin(pricy_lowV.index)].groupby('Name')['date'].agg(['min', 'max']).reset_index()
ph_date_range = stock_df[stock_df['Name'].isin(pricy_hiV.index)].groupby('Name')['date'].agg(['min', 'max']).reset_index()
pl_date_range.head()
```

```python
def percent_gain_dict(stk_names, date_range_df):

    pg_dict = {}
    for s in stk_names:   
        min_date = date_range_df[date_range_df['Name'] == s]['min']
        max_date = date_range_df[date_range_df['Name'] == s]['max']
        s_begin = float(stock_df[(stock_df['Name'] == s) & (stock_df['date'].isin(min_date))]['close'])
        s_end = float(stock_df[(stock_df['Name'] == s) & (stock_df['date'].isin(max_date))]['close'])
        pg_dict[s] = (s_end - s_begin) / s_begin

    return pg_dict
```

```python
# 'NVDA' is the pricy stock with relatively low volatility and highest percentage gain over 5 yrs
ph_dict = percent_gain_dict(pricy_hiV.index, ph_date_range)

sorted_ph_dict = sorted(ph_dict.items(), key=lambda x: x[1], reverse=True)
sorted_ph_dict
```
